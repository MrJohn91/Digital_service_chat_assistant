#### Spotify FAQ and Support

```python
import pandas as pd
```

```python
import requests

# Fetch the webpage content
url = 'https://support.spotify.com/us/'
response = requests.get(url)

# Get HTML content
html_content = response.text

# Save to a file
with open('spotify_support.html', 'w') as file:
    file.write(html_content)

print("HTML content saved to spotify_support.html")
```

```python
# check  if request was successful
if response.status_code == 200:
    print("Successfully fetched the webpage content!")
```

```python
# check the length of the HTML content
print(f"Length of HTML content: {len(html_content)} characters")
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_content, 'html.parser')

# Find all potential questions
questions = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'li'])

for question in questions:
    print(question.text)
```

```python
# find heding tags
for heading in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5']):
    question = heading.get_text(strip=True)
    answer = ""
    
    # Find the next sibling that contains the answer
    sibling = heading.find_next_sibling()
    if sibling:
        answer = sibling.get_text(strip=True)
    
    print(f"Question: {question}")
    print(f"Answer: {answer}")
    print("-" * 50)
```

#### Scrapping the plan categroy

```python
url = 'https://community.spotify.com/t5/FAQs/tkb-p/Spotify-Answers/label-name/plan%20help'
response = requests.get(url)
html_content = response.text

soup = BeautifulSoup(html_content, 'html.parser')
faq_items = soup.find_all('article', class_='custom-faq-tile')

faq_data = []

for faq in faq_items:
    question_button = faq.find('button', class_='faq-question')
    question = question_button.text.strip() if question_button else 'No question found'
    
    answer_div = faq.find('div', class_='faq-answer')
    answer = answer_div.get_text(separator=" ", strip=True) if answer_div else 'No answer found'
    
    faq_data.append({
        'question': question,
        'answer': answer
    })
for item in faq_data:
    print(f"Question: {item['question']}")
    print(f"Answer: {item['answer']}")
    print("-" * 50)
```

#### All Categories

```python
# list of all category URLs
urls = [
    'https://community.spotify.com/t5/FAQs/tkb-p/Spotify-Answers/label-name/app%20help',
    'https://community.spotify.com/t5/FAQs/tkb-p/Spotify-Answers/label-name/plan%20help',
    'https://community.spotify.com/t5/FAQs/tkb-p/Spotify-Answers/label-name/%20payment%20help',
    'https://community.spotify.com/t5/FAQs/tkb-p/Spotify-Answers/label-name/%20device%20help',
    'https://community.spotify.com/t5/FAQs/tkb-p/Spotify-Answers/label-name/%20account%20help',
    'https://community.spotify.com/t5/FAQs/tkb-p/Spotify-Answers/label-name/%20community%20help'
]

all_faq_data = []

for url in urls:
    response = requests.get(url)
    html_content = response.text

    soup = BeautifulSoup(html_content, 'html.parser')

    faq_items = soup.find_all('article', class_='custom-faq-tile')

    for faq in faq_items:
        # Extract question
        question_button = faq.find('button', class_='faq-question')
        question = question_button.text.strip() if question_button else 'No question found'

        # Extract answer
        answer_div = faq.find('div', class_='faq-answer')
        answer = answer_div.get_text(separator=" ", strip=True) if answer_div else 'No answer found'

        # Append to all_faq_data
        all_faq_data.append({
            'question': question,
            'answer': answer
        })
for item in all_faq_data:
    print(f"Question: {item['question']}")
    print(f"Answer: {item['answer']}")
    print("-" * 50)
```

```python
# json library to save the data
import json
# Save the FAQ data to a JSON file
with open('spotify_faq_data.json', 'w') as json_file:
    json.dump(all_faq_data, json_file, indent=4)
```

```python
# load the json file
with open('spotify_faq_data.json', 'r') as json_file:
    loaded_faq_data = json.load(json_file)
    # verify the loaded data
    for item in loaded_faq_data:
        print(f"Question: {item['question']}")
        print(f"Answer: {item['answer']}")
        print("-" * 50)
```

### Embeddings 

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
# Create an embedding model to convert text to vectors
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Prepare the FAQ text for embedding
faq_text = [f"{faq['question']} {faq['answer']}" for faq in faq_data]

vector_store = FAISS.from_texts(faq_text, embedding_model)
```

```python
# function to search faqs
def get_relevant_faqs(query):
    return vector_store.similarity_search(query, k=3)
# test 
user_query = "How can I update my family plan?"
results = get_relevant_faqs(user_query)

for result in results:
    print(result.page_content)
```

```python
from langchain.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,  
    chain_type="stuff",  
    retriever=vector_store.as_retriever()
)
# test the QA chain
query = "How can I check the address Iâ€™ve registered on my Premium Family account?"
result = qa_chain.invoke({"query": query})
print(result)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings, HuggingFaceEndpoint
from langchain.vectorstores import FAISS
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

faq_text = [f"{faq['question']} {faq['answer']}" for faq in faq_data]

# faiss vector
vector_store = FAISS.from_texts(faq_text, embedding_model)

hf_model = "mistralai/Mistral-7B-Instruct-v0.3"
llm = HuggingFaceEndpoint(repo_id=hf_model)

memory = ConversationBufferMemory(memory_key="chat_history")
# conversational retrieval chain
qa_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=vector_store.as_retriever(), # vector store as retriever
    memory=memory,
    verbose=True
)

def ask_question(query):
    result = qa_chain({"question": query})
    return result["answer"]
# test the function
user_query = "How can I update my family plan?"
response = ask_question(user_query)
print(response)
```
